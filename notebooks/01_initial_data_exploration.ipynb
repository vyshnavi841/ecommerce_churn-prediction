```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
# Load the raw dataset
df = pd.read_csv('../data/raw/online_retail.csv')
print(f"Dataset Shape: {df.shape}")
df.head(10)
```

```python
# Q1: What is the date range of transactions?
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print(f"Date Range: {df['InvoiceDate'].min()} to {df['InvoiceDate'].max()}")
```

```python
# Q2: How many unique customers?
print(f"Unique Customers: {df['CustomerID'].nunique()}")
```

```python
# Q3: How many unique products?
print(f"Unique Products: {df['StockCode'].nunique()}")
```

```python
# Q4: How many countries represented?
print(f"Countries represented: {df['Country'].nunique()}")
```

```python
# Q5: How many missing values in each column?
df.isnull().sum()
```

```python
# Q6: What percentage of CustomerID is missing?
missing_pct = (df['CustomerID'].isnull().sum() / len(df)) * 100
print(f"Missing CustomerID Percentage: {missing_pct:.2f}%")
```

```python
# Q7: Are there any duplicate invoices?
print(f"Duplicate Rows: {df.duplicated().sum()}")
```

```python
# Q8: What's the distribution of Quantity?
df['Quantity'].describe()
```

```python
# Q9: Are there negative quantities? (Returns/Cancellations)
print(f"Negative Quantities: {(df['Quantity'] < 0).sum()}")
```

```python
# Q10: What's the price range (min, max, median)?
print(f"Price Range: Min = {df['UnitPrice'].min()}, Max = {df['UnitPrice'].max()}, Median = {df['UnitPrice'].median()}")
```

```python
# Q11: Are there any unusual InvoiceNo patterns?
cancelled = df['InvoiceNo'].astype(str).str.startswith('C').sum()
print(f"Cancelled Invoices (Start with 'C'): {cancelled}")
```

```python
# Q12: Any zero or negative prices?
print(f"Zero or Negative Prices: {(df['UnitPrice'] <= 0).sum()}")
```

```python
# Q13: Any extremely high quantities (potential data errors)?
print(f"Extremely high quantities (> 1000): {(df['Quantity'] > 1000).sum()}")
```

```python
# Data Quality Summary
data_quality_summary = {
    'total_rows': len(df),
    'total_columns': len(df.columns),
    'missing_values': df.isnull().sum().to_dict(),
    'duplicate_rows': int(df.duplicated().sum()),
    'date_range': {
        'start': str(df['InvoiceDate'].min().date()),
        'end': str(df['InvoiceDate'].max().date())
    },
    'negative_quantities': int((df['Quantity'] < 0).sum()),
    'cancelled_invoices': int(df['InvoiceNo'].astype(str).str.startswith('C').sum()),
    'missing_customer_ids': int(df['CustomerID'].isnull().sum()),
    'missing_customer_ids_percentage': float((df['CustomerID'].isnull().sum() / len(df)) * 100)
}

# Save to JSON
import json
import os
os.makedirs('../data/raw', exist_ok=True)
with open('../data/raw/data_quality_summary.json', 'w') as f:
    json.dump(data_quality_summary, f, indent=4, default=str)

print("Saved data_quality_summary.json")
```
